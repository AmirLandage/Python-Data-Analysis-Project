Change the index to the release date 

```python
df_tracks.set_index("release_date", inplace = True)
df_tracks.index = pd.to_datetime(df_tracks.index)
df_tracks.head()
```

Check the artist in the 18th row of the data set

```python
df_tracks[['artists']].iloc[18]
```

Convert the duration of miliseconds into seconds

```python
df_tracks["duration"] = df_tracks["duration_ms"].apply(lambda x: round(x/1000))
df_tracks.drop("duration_ms", inplace = True, axis = 1)
df_tracks.head()
```

Create a heatmap to show the correlation among the comumns

```python
corr_df = df_tracks.drop(['key','mode','explicit'], axis = 1).corr(method = 'pearson')
plt.figure(figsize = (14,6))
heatmap = sns.heatmap(corr_df, annot = True,fmt = ".1g", vmin = -1, vmax = 1,center = 0, cmap = 'inferno', linewidths = 1, linecolor = 'Black')
heatmap.set_title("Correlation Heatmap Between Variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation = 90)
```

create a regression plot between loudness and energy

```python
plt.figure(figsize=(10,6))
regplot = sns.regplot(data = sample_df, y = "loudness", x = "energy", color = 'c')
regplot.set_title("Loudness vs Energy Correlation")
```

create a regression plot between popularity and danceability

```python
plt.figure(figsize=(10,6))
regplot_pd = sns.regplot(data = sample_df, y = 'popularity', x = 'danceability', color ='b')
regplot_pd.set_title("Popularity vs Danceability")
```

Ceate a Year column form the release date column

```python
df_tracks["dates"] = df_tracks.index.get_level_values('release_date')
df_tracks.dates = pd.to_datetime(df_tracks.dates)
years = df_tracks.dates.dt.year
df_tracks.head()
```

Create a distribution plot to visualize total number of songs each year scince 1922 on Spotify

Create a line plot to analyze average duration of songs over the years

```python
total_dr = df_tracks.duration
sns.set_style(style= 'whitegrid')
fig_dims = (15,8)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.lineplot(x = years, y = total_dr, ax = ax).set(title = "Year vs Duration")
plt.xticks(rotation = 60)
```
